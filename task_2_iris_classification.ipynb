## Iris flower classification

#### Importing libraries

```python
# data analysis
import numpy as np
import pandas as pd

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# dataset
from sklearn.datasets import load_iris

# train-test split
from sklearn.model_selection import train_test_split

# models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# evaluation metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

#### Loading dataset

```python
# loading dataset
iris = load_iris()

X = iris.data
y = iris.target
```

```python
# creating dataframe
iris_df = load_iris(as_frame=True).frame
```

#### Data Visualization

```python
# bar chart
species_counts = iris_df["target"].value_counts()

plt.figure(figsize=(6,4))
species_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])

plt.title("Number of Samples in Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.xticks(rotation=0)

plt.show()
```

```python
# histogram
plt.figure(figsize=(6,4))

plt.hist(iris_df["petal length (cm)"],
         bins=15,
         color="orange",
         edgecolor="black")

plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

```python
# correlation matrix 
corr = iris_df.corr()

# visualizing correlation matrix using heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# visualizing pairplot (all the scatter plots between two features)
sns.pairplot(iris_df, hue='target')
plt.show()
```

For classification, we will use all the four features to classify to have best accuracy predictions. Though, petal length and petal width impart the most information as they show less overlap between species than sepal measurements.

#### Conclusion

The Iris dataset was successfully explored using various data visualization techniques. The bar chart showed that all three Iris species are equally represented in the dataset. The histogram illustrated the distribution of numerical features, helping to understand their spread and frequency. The scatter plot revealed clear relationships between different features, particularly showing that petal measurements are effective in distinguishing between species. The pair plot and correlation heatmap further highlighted the relationships among features, indicating strong correlations between petal length and petal width. Overall, these visualizations provide valuable insights into the structure of the Iris dataset and demonstrate that it is well-suited for exploratory data analysis and classification tasks.
